```python
# All null values are '[]', so therefore I create function that fills all these values with 'Null'
def fill_null(x):
    return x.replace('[]', 'Null')

# A function to get rid of brackets around are data points.
def rid_brackets(x):
    return x.replace("['","").replace("']", "")
    

#Returns the last 4-7 digits, which is the article number    
def prod_nr(x):
    import re
    return re.findall(r'\d{4,7}\b', x)

# returns 1 or 2 digits
def gram_liter(x):
    import re
    return re.findall(r'\d{1,2}\b', x)

def volym(x):
    return float(x.replace(' ml', '').replace(',','.').replace('Null', '0'))

def procent(x):
    return float(x.replace(' %', '').replace(',','.').replace('Null', '0'))


   
```

```python
def sub_(x):
    import re
    return re.sub(r'\{\d*}', '', x)
```

```python
df = df.applymap(fill_null)
```

```python
df = df.applymap(rid_brackets)
```

```python
df['Produktnummer'] = df['Produktnummer'].apply(prod_nr)
```

```python
df['Sockerhalt(gram/l)'] = df['Sockerhalt(gram/l)'].apply(gram_liter)
```

```python
df['Volym(mml)'] = df['Volym(mml)'].apply(volym)
```

```python
df['Alkoholhalt(%)'] = df['Alkoholhalt(%)'].apply(procent)
```

```python
df.drop(columns='Doft', axis=1, inplace=True)
```

```python


def doft(x):
    import re
    return re.findall(r'Doft(.+?)[.]', x)
    
def color(x):
    import re
    return re.findall(r'Färg(.+?)[.]', x)
    
def ravaror(x):
    import re
    return re.findall(r'Råvaror(.+?)[.]', x)
    
def smaken(x):
    import re
    return re.findall(r'Smakbeskrivning(.+?)[.]', x)

def year(x):
    import re
    x = re.findall(r'\d{4}\b', x)
    return x

def fixprice1(x):
    x = x.replace(',','')
    return x

def fixprice2(x):
    length = len(x)
    first_half = x[:int(len(x)/2)]
    return first_half

def fixprice3(x):
     return x.replace(" ","").replace(":-'", "")
    
```

```python
df['Doft'] = df['rawdata1'].apply(doft)
```

```python
df['Färg'] = df['rawdata1'].apply(color)
```

```python
df['Råvaror'] = df['rawdata1'].apply(ravaror)
```

```python
df['Smakbeskrivning'] = df['rawdata1'].apply(smaken)
```

```python
df['Årgång'] = df['Årgång'].apply(year)
```

```python
df['Pris(SEK)'] = df['Pris(SEK)'].apply(fixprice1)
```

```python
df['Pris(SEK)'] = df['Pris(SEK)'].apply(fixprice2)
```

```python
df['Pris(SEK)'] = df['Pris(SEK)'].apply(fixprice3)
```

```python
def check(checked,x) :
    a = 0
    if len(x)>0:
        if checked in x:
            a=1
    return a
```

```python
# Creating dummy varibale columns for all the types of food the wine is a good fit to.

list_checked = ['Lamm', 'Nöt', 'Fågel', 'Fläsk', 'Dessert','Vilt','Ost']

for checked in list_checked :
    df[checked] = df['passar_till'].apply(lambda x: check(checked,x))
```

```python
def lowerlst(x):
    x = [i.lower() for i in x]
    return x
```
